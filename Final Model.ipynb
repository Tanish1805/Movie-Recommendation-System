```python
#Getting Overview Column
from tmdbv3api import Movie
tmdb_movie = Movie()
def get_genre(x):
    genres = []
    #searching the title of the movie in tmdb
    result = tmdb_movie.search(x)
    #checks if the movie is present in TMDB database
    if(result):
        
        movie_id = result[0].id
        #requesting to accesss the info of that movie title
        response = requests.get('https://api.themoviedb.org/3/movie/{}?api_key={}'.format(movie_id,tmdb.api_key))
        #converting it into jason file
        data_json = response.json()
        #getting genres column and appending in our data
        if data_json['overview']:
            genre_str = " " 
            genres.append(data_json['overview'])
            return genre_str.join(genres)
    else:
        genre_str = "not found"
        return genre_str
```

```python
#Getting Taglines Column
from tmdbv3api import Movie
tmdb_movie = Movie()
def get_tagline(x):
    taglines = []
    #searching the title of the movie in tmdb
    result = tmdb_movie.search(x)
    #checks if the movie is present in TMDB database
    if(result):
        
        movie_id = result[0].id
        #requesting to accesss the info of that movie title
        response = requests.get('https://api.themoviedb.org/3/movie/{}?api_key={}'.format(movie_id,tmdb.api_key))
        #converting it into jason file
        data_json = response.json()
        #getting genres column and appending in our data
        if data_json['tagline']:
            tagline_str = " " 
            taglines.append(data_json['tagline'])
            return tagline_str.join(taglines)
    else:
        tagline_str = "not found"
        return tagline_str
```

```python
#Getting Rating Column
from tmdbv3api import Movie
tmdb_movie = Movie()
def get_rating(x):
    rating = []
    #searching the title of the movie in tmdb
    result = tmdb_movie.search(x)
    #checks if the movie is present in TMDB database
    if(result):
        
        movie_id = result[0].id
        #requesting to accesss the info of that movie title
        response = requests.get('https://api.themoviedb.org/3/movie/{}?api_key={}'.format(movie_id,tmdb.api_key))
        #converting it into jason file
        data_json = response.json()
        #getting genres column and appending in our data
        if data_json['vote_average']:
            rating_str = " " 
            rating.append(str(data_json['vote_average'])) #the data was float type so we convert to string
            return rating_str.join(rating)
    else:
        rating_str = "not found"
        return rating_str
```

```python
#Getting production companies Column
from tmdbv3api import Movie
tmdb_movie = Movie()
def get_production(x):
    company = []
    #searching the title of the movie in tmdb
    result = tmdb_movie.search(x)
    #checks if the movie is present in TMDB database
    if(result):
        
        movie_id = result[0].id
        #requesting to accesss the info of that movie title
        response = requests.get('https://api.themoviedb.org/3/movie/{}?api_key={}'.format(movie_id,tmdb.api_key))
        #converting it into jason file
        data_json = response.json()
        #getting genres column and appending in our data
        if data_json['production_companies']:
            company_str = " " 
            for i in range(0,len(data_json['production_companies'])):
              company.append(data_json['production_companies'][i]['name'])
            return company_str.join(company)
    else:
        company_str = "not found"
        return company_str
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfv=TfidfVectorizer(min_df=3, max_features=None, 
                    strip_accents='unicode', analyzer='word', token_pattern=r'\w{1,}', 
                    ngram_range=(1,3), stop_words='english')
```

```python
#creating a function to recommend movie according to the sorted list of cosine similarity of that movie
def recommend(movie):
    movie_index = new_df[new_df['title_nospace'] == movie].index[0]
    distances = similarity[movie_index]
    #getting top 10 recommended movies
    movies_list = sorted(list(enumerate(distances)), reverse=True , key=lambda x:x[1])[1:11]
    for i in movies_list:
        recommended_movie_index = new_df.iloc[i[0]]
        print("Title:", recommended_movie_index['movie_title'])
        print("Rating:", recommended_movie_index['rating'])
        print("Overview:", recommended_movie_index['overview'])
        print()
```

# Predicting the Movies with the use of Collaborative Filtering

```python
rating_popular_movie = pd.read_csv('rating_popular_movie.csv', usecols = ['userId' ,'movieId', 'rating' , 'totalratingcount', 'movie_name', 'movie_name_nospace'], dtype ={'userId': 'int32', 'movieId': 'int32', 'rating': 'float32', 'totalratingcount': 'int32'})
rating_popular_movie
```

```python
## First lets create a Pivot matrix
## see below at userId=3 and title = A Space Odyssey the value is 5 
## means user 3rd has given 5 rating to A Space Odyssey
movie_features_df=rating_popular_movie.pivot_table(index='movie_name_nospace',columns='userId',values='rating').fillna(0)
movie_features_df
```

```python
## Creating an array matrix of the pivot table 
from scipy.sparse import csr_matrix

movie_features_df_matrix = csr_matrix(movie_features_df.values)
```

```python
#Using Nearest Neighbours(unsupervised ML) we will use cosine similarity
# Explained in detal in copy
from sklearn.neighbors import NearestNeighbors


model_knn = NearestNeighbors(metric = 'cosine', algorithm = 'brute')
model_knn.fit(movie_features_df_matrix)
```

```python
movie_features_df.shape
```

```python
movie_features_df.shape[0]
```

```python
# if the movie exists in this pivot table find its index
def get_movie_index(movie_name, pivot):
    try:
        movie_index = pivot.index.get_loc(movie_name)
        return movie_index
    except KeyError:
        return None
```

```python
#input the user can give
movie_name = "tenet"
index = get_movie_index(movie_name, movie_features_df)
print(index)
```

```python
#Getting distances and index and predecting recommended movies acc to user to user model
if(index):
    distances, indices = model_knn.kneighbors(movie_features_df.iloc[index,:].values.reshape(1, -1), n_neighbors = 6)
    for i in range(0, len(distances.flatten())):
        if i == 0:
            print('Recommendations for {0}:\n'.format(movie_features_df.index[index]))
        else:
            print('{0}: {1}'.format(i, movie_features_df.index[indices.flatten()[i]]))
```
